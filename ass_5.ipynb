```python
import pandas as pd
import numpy as np
path="/content/titanic_dataset.csv"
df=pd.read_csv(path)
df.head()
```

```python
#2. Set 'PassengerId' as the index column
df.set_index('PassengerId', inplace=True)

# Print the updated DataFrame
print(df.head())
```

```python
#4. Fill in all the missing values present in all the columns in the dataset
df_filled = df.fillna(0)

# Print the updated DataFrame
print(df_filled.head())
```

```python
#5. Check and handle outliers in at least 3 columns in the dataset
columns_to_check = ['Age', 'Fare', 'SibSp']

# Calculate z-scores for the selected columns
z_scores = np.abs((df[columns_to_check] - df[columns_to_check].mean()) / df[columns_to_check].std())

# Define a threshold for identifying outliers
threshold = 3

# Identify outliers by comparing z-scores to the threshold
outliers = (z_scores > threshold).any(axis=1)

# Handle outliers by replacing them with the median of the column
df.loc[outliers, columns_to_check] = df[columns_to_check].median()

# Print the updated DataFrame
print(df[columns_to_check].head())

```

```python
#6.Do min max scaling on the feature set (Take ‘Survived’ as target)
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Separate the feature set (X) and the target variable (y)
X = df.drop('Survived', axis=1)  # Features
y = df['Survived']  # Target

# Identify numeric and categorical columns
numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Define column transformers for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('numeric', MinMaxScaler(), numeric_cols),
        ('categorical', OneHotEncoder(), categorical_cols)
    ])

# Apply preprocessing to feature set
X_scaled = preprocessor.fit_transform(X)

# Print the scaled feature set
print(X_scaled)

```
